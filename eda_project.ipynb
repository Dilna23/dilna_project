```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

Importing dataset

```python
df=pd.read_csv("/content/Books_Data_Clean (1).csv")

print(df)
```

Data Inspection: Investigating Negative Publishing Years

```python
df[df["Publishing Year"] < 0].head()
```

```python
df = df[df["Publishing Year"] >= 1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset="Book Name",inplace=True)
```

```python
df.nunique()
```

Visualization


```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.show()
```

```python
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of Books in each Genre")
plt.show()
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False)
```

```python
top_authors = df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False).head(10)


top_authors.plot(kind="barh")
plt.xlabel("Average Rating")
plt.ylabel("Author")
plt.title("Top 10 Authors by Average Rating")
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.scatter(df["sale price"],df["units sold"])
plt.xlabel("Sale price")
plt.ylabel("Units sold")
plt.title("Scatter plot of Sale Price vs Unit Sold")
plt.show()
```

```python
sns.boxplot(x="Author_Rating",y="units sold", data=df)
plt.xlabel("Auhor Rating")
plt.ylabel("Units Sold")
plt.title("Boxplot of unit sold for each author rating")
plt.show()
```

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line",marker="o")
plt.xlabel("Publish Year")
plt.ylabel("Units Sold")
plt.title("Total units sold over the years")
plt.show()
```
